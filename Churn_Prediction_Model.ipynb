```python
import pandas as pd
```

```python
# Load the dataset
df = pd.read_csv('/content/WA_Fn-UseC_-Telco-Customer-Churn.csv')
df.head()
```

```python
df.info()
```

```python
# Check for missing values
print("\nMissing values:")
print(df.isnull().sum())
```

```python
# Check data types of columns
print("\nData types:")
print(df.dtypes)
```

```python
# Encode categorical variables
df_encoded = pd.get_dummies(df)
```

```python
# Split the dataset into features and target variable
X = df_encoded.drop('Churn_Yes', axis=1)
y = df_encoded['Churn_Yes']
```

# Model Development and Evaluation

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Create a Logistic Regression model
model = LogisticRegression()
```

```python
# Train the model
model.fit(X_train, y_train)
```

```python
# Make predictions on the testing set
y_pred = model.predict(X_test)
```

```python
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

# Print the evaluation metrics
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1-Score:", f1)
```

This results suggest that the model is predicting all the churn instances correctly, resulting in no false negatives or false positives. <br/>
While achieving perfect scores on all evaluation metrics may seem desirable, it could also indicate a potential issue with the model or the dataset. Here are a few considerations: <br/>
1. Imbalanced Dataset: Check if the dataset is imbalanced, meaning that one class (e.g., churn) is significantly underrepresented compared to the other class (e.g., non-churn). In such cases, a model that predicts the majority class for all instances can still achieve high accuracy, but it may not be useful in practice. To address this, you can explore techniques like oversampling the minority class or using different evaluation metrics, such as precision, recall, and F1-score.

2. Data Leakage: Ensure that the dataset has been properly split into training and testing sets, and there is no data leakage between them. Data leakage occurs when information from the testing set unintentionally influences the training process, leading to overly optimistic performance metrics. Double-check your data splitting process to rule out data leakage as a cause for the high scores.

3. Model Overfitting: Another possibility is that the model is overfitting the training data, meaning it is learning the specific patterns in the training set too well and failing to generalize to new, unseen data. To confirm this, you can evaluate the model's performance on a separate validation set or perform cross-validation. If the model's performance on the validation set or during cross-validation is significantly lower than on the training set, it indicates overfitting.
